```python
import gymnasium as gym
import numpy as np
import pandas as pd
import random
```

#### Sample Environments 

```python
env=gym.make("CliffWalking-v1")
```

```python
print(env.observation_space.n) #states
print(env.action_space.n) #actions
starting_point,_=env.reset() #36=13*12+0
print(starting_point) #starting point
```

## SARSA

```python
# define Parameters
gamma=0.99
alpha=0.5
epsilon=0.1
episodes=500
#Q-Table
Q=np.zeros((48,4))
```

```python
#Policy epsilon -Greeedy : state -> action state base action
def epsilon_greedy(state):
    if random.random()>epsilon:
        return env.action_space.sample() #explore the state
    else:
        return np.argmax(Q[state])
```

### Training Agent

```python
for episode in range(episodes):
    env=gym.make("CliffWalking-v1")
    done=False
    state,_=env.reset()
    action=epsilon_greedy(state)

    total_reward=0
    epsiode_len=0
    while not done:
        next_state,reward,terminated,truncated,_=env.step(action)
        done=terminated or truncated
        next_action=epsilon_greedy(next_state)

        # SARSA UPDATION  FORMULA
        Q[state,action] += alpha * (reward + gamma * (Q[next_state, next_action] - Q[state, action]))

        state=next_state
        action=next_action

        total_reward+=reward
        epsiode_len+=1
    print(f"episode={episode+1}/{episodes} and total rewards={total_reward} and episode length={epsiode_len}")
        
    
```
